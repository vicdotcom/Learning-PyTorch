### Reshaping, Stacking, Squeezing & Unsqueezing Tensors
- Reshaping- Reshapes an input tensor to a defined shape
- View- Return a view of an input tensor of a certain shape but keep the same memory as the original tensor (we can 'view' how the tensor would look like if it were reshaped and make changes to it to observe how it would affect the original tensor)
- Stacking- Combine multiple tensors on top of each other (using `torch.stack()`)
   - There is also concatenation (using `torch.cat()`)
- Squeezing- Removes all '1' dimensions from a tensor
- Unsqueezing- Adds a '1' dimension to a tensor
- Permute- Returns a view of the input tensor with dimensions permuted (swapped) in a diffreent way

````python
import torch
import numpy as np
````

````python
z= torch.arange(20)

z.shape, z
 # Vector shape (1 dimension)
````

#### Reshape

````python
## Add an extra column dimension
 # The below code shows we want to shape a tensor from a 1D vector to a 2D matrix
 # The matrix should have 1 row and 'n' columns
 # 'n' depends on the length of the 1D matrix, otherwise, reshaping will bring an error
z_reshaped= z.reshape(1, len(z))
 
z_reshaped.shape, z_reshaped
````

````python
## Add extra row dimensions
 # The rule of 'n' still applies
z_reshaped_row= z.reshape(len(z), 1)
z_reshaped_row.shape, z_reshaped_row
````

````python
## Basically, the new reshaped tensor should be able to hold the....
## ...same no. of elements as the original tensor
z_shaped= z.reshape(4,5)
z_shaped1= z.reshape(10,2)
print(z_shaped.shape)
print(z_shaped)
print()
print(z_shaped1.shape)
print(z_shaped1)


````

#### Change View

````python
z
````

````python
## Changing the view
v= z.view(4,5)
v.shape, v
````

````python
## Changing the viewing tensor 'v' also changes the orignal tensor z
## This is because the view of a tensor shares the same memory as the original tensor
## Here we change the element on the first row of the third column
v[1,3]= 300
v, z
````

#### Stacking
Stacking combines two or more tensors on top of or beside each other using `torch.stack()`.

````python
z.shape, z
````

````python
z_stacked= torch.stack([z, z])
z_stacked.shape, z_stacked
 # The individual vector tensors are stacked on top of each other to create a mattrix tensor
 # New (row) dimension is added here
````

````python
## We can specify a target dimension we want the stacked tensor to appear in
 # Default is dim= 0 (row stacking)
 # dim= 1 peforms column stacking
z_stacked1= torch.stack([z,z], dim= 1)
z_stacked1.shape, z_stacked1
````

While stacking *stacks* whole tensors, concatenation combines *the elements of a tensor* along a given dimension using `torch.cat()`.  

````python
z_concat= torch.cat((z,z), dim= 0) # Concatenation along rows
print(z_concat, z_concat.shape, z_concat.ndim)
 # 
````

So unlike in stacking where we get a new tensor shape, in concatenation the original tensor shape and dimension is retained.

Concatenation can only occur along an existing dimension meaning in the above example, we cannot concatenate column-wise (there are no columns as the original tensor `z` in the above example was a 1D vector)

But if we were dealing with a matrix, this would be possible.

````python
zz= z.reshape(1, -1)
print(f'Original Tensor: {zz.shape, zz.ndim}')


zz_concat= torch.concat((zz, zz), dim= 0)
print(f'\n Concatenated Tensor: {zz_concat.shape, zz_concat.ndim, zz_concat}')
````

#### Squeezing
Removes all dimensions of size 1 from the input tensor.

````python
## Original tensor
s= torch.rand([1,15])
print('Original tensor and tensor shape')
print(s.shape, s)
print('\nSqueezed tensor and tensor shape')
print(s.squeeze().shape, s.squeeze())
````

````python
s1= torch.rand([4, 5, 1, 3, 1])
 # 5 dimensional tensor
print('Original tensor shape')
print(s1.shape)
print('\nSqueezed tensor shape')
print(s1.squeeze().shape)
````

#### Unsqueeze
Adds a dimension of size 1 to the tensor's shape.

````python
s_squeezed= s1.squeeze()
print('Original Tensor shape')
print(s_squeezed.shape)

print('\nNew tesnor shape (dim= 0)')
print(s_squeezed.unsqueeze(dim= 0).shape)

print('\nUnsqueezed tensor shape (dim= 1)')
print(s_squeezed.unsqueeze(dim= 1).shape)

print('\nUnsqueezed tensor shape (dim= 1)')
print(s_squeezed.unsqueeze(dim= 1).shape)

print('\nUnsqueezed tensor shape (dim= 3)')
print(s_squeezed.unsqueeze(dim= 3).shape)
````

#### Permute
- Can be used to change (swap) the order of the dimensions of a tensor.
- Common when working with image tensors

````python
## We create an image tensor [height, width, color channels]
p= torch.rand([220, 221, 3])
p_permuted= p.permute(2, # dim 2 is mapped/swapped to the first index
                      0, # dim 0 is swapped to the 2nd index
                      1) # dim 1 is swapped to the last/3rd index

print(f'Previous shape {p.shape}')
print(f'New shape: {p_permuted.shape}')

print('\nChecking whether changes are made')
print(f'Original: {p[0,0,2]}') # Row 0, column 0, matrix 2
print(f'Permuted: {p_permuted[0,0,2]}')
````

### Replacing elements in a tensor
This can be done using `torch.where()` which takes a condition and replaces values in a tensor based on that condition.
```python
torch.where(condition, x, y) 
 # torch.where(condition, replacement if condition is True, otherwise if False)
 ```


````python
a, b= -1, 1
rand_tensor= a + (b - a) * torch.rand(4,4)
print(f'Original Tensor: {rand_tensor, rand_tensor.shape, rand_tensor.dtype}')

## Replacing values less than 0
rand_tensor_replaced= torch.where(rand_tensor < 0, # Condition
                                  torch.tensor(0), # Replace with value 0 (We use a scalar tensor to represent 0 to ensure the same float32 datatype)
                                  rand_tensor) # Otherwise
print(f'\nReplaced Tensor: {rand_tensor_replaced, rand_tensor_replaced.shape, rand_tensor_replaced.dtype}')
````

### PyTorch Tensors & NumPy
NumPy is a popular scientific numerical computing library. Because of this, PyTorch has the functionality to interact with it.
- If the data is in NumPy and we want it in PyTorch -> `torch.from_numpy(array_data)`
- If the data is in PyTorch and we want it in NumPy -> `tensor_data.numpy()`

They do share memory meaning changing one will change the other.

````python
## NumPy array to tensor
array= np.random.random([3,3])
print('Array')
print(array)

tensor= torch.from_numpy(array)
print('\nTensor')
print(tensor)
 # Datatype of the converted tensor is float64 as is the default datatype for NumPy arrays
 # Default datatype for PyTorch tensors is float32
````

````python
## Tensor to NumPy array
print('Tensor')
print(tensor)

print('\nArray')
print(torch.Tensor.numpy(tensor))
````

````python
## Original tensor
tensor14= torch.tensor([[3,3,3], [3,3,3], [3,3,3]])
print(f'Original Tensor: {tensor14, tensor14.shape, tensor14.device}')

# Conversion of the tensor to an array
array14= torch.Tensor.numpy(tensor14)
print(f'\nOriginal Numpy Array: {array14, array14.shape}')

## Multiplying array elements by 2
array14 *= 2
print(f'\nNew Numpy Array: {array14, array14.shape}')

print(f'\nChecking the Tensor: {tensor14, tensor14.shape}')

## The original tensor changes as well meaning that they share memory
````
